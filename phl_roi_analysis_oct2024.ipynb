```python
# ROI Simulator for Cold Storage in Nigeria (2024 Price Data)
import pandas as pd

def calculate_roi(crop, volume_tons, harvest_price, peak_price, energy_cost_monthly):
    # Constants
    waste_ambient = 0.40  # 40% loss
    waste_cold = 0.05     # 5% loss
    months_stored = 4

    # 1. Status Quo (No Storage)
    marketable_vol_no_cold = volume_tons * (1 - waste_ambient)
    revenue_no_cold = marketable_vol_no_cold * peak_price

    # 2. Cold Chain (With Storage)
    marketable_vol_cold = volume_tons * (1 - waste_cold)
    total_energy_cost = energy_cost_monthly * months_stored
    revenue_cold = (marketable_vol_cold * peak_price) - total_energy_cost

    # 3. Results
    profit_boost = revenue_cold - revenue_no_cold
    roi_percent = (profit_boost / (harvest_price * volume_tons)) * 100

    print(f"--- Analysis for {crop} ---")
    print(f"Revenue saved from rot: ₦{profit_boost:,.2f}")
    print(f"Projected ROI: {roi_percent:.2f}%")

# Using the NBS Oct 2024 Price Trends
calculate_roi("Irish Potatoes", volume_tons=10, harvest_price=850000, peak_price=1600000, energy_cost_monthly=150000)
```
